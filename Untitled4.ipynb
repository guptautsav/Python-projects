#Customer Lifetime value analysis project

```python
import pandas as pd
import plotly.graph_objects as go
import plotly.express as px
import plotly.io as pio
pio.templates.default = "plotly_white"
```

```python
data = pd.read_csv("customer_acquisition_data.csv")
```

```python
data
```

#Visualizing the distribution of acquisition cost and revenue generated by the customer usig histogram.

```python
fig = px.histogram(data, x="cost",nbins=20, title="Distribution of Acquisition Cost")
fig.show()
```

```python
fig = px.histogram(data, x="revenue",nbins=20, title="Distribution of Acquisition Cost")
fig.show()
```

```python
cost_by_channel = data.groupby("channel")["cost"].mean().reset_index()
revenue_by_channel = data.groupby("channel")["revenue"].sum()
```

```python
cost_by_channel
```

```python
cost_by_channel = data.groupby("channel")["cost"].mean().reset_index()

fig = px.bar(cost_by_channel,
             x="channel", y="cost",
             title="Average Acquisition Cost by Channel")
fig.show()
```

#Conversion rate by Channel

```python
data.head()
```

```python
conversion_by_channel = data.groupby("channel")["conversion_rate"].mean().reset_index()


```

```python
conversion_by_channel
```

```python
conversion_by_channel = data.groupby("channel")["conversion_rate"].mean().reset_index()

fig = px.bar(conversion_by_channel,
             x="channel", y="conversion_rate",
             title="Conversion Rate by Channel")
fig.show()
```

#Total revenue by channel

```python
revenue_by_channel = data.groupby('channel')['revenue'].sum().reset_index()

fig = px.pie(revenue_by_channel,
             values="revenue", names="channel",
             title="Total revenue by Channel",
             hole=0.6, color_discrete_sequence=px.colors.qualitative.Pastel)

fig.show()
```

#Return on Investment for each channel

```python
data['roi'] = (data['revenue']) / data['cost']
roi_by_channel = data.groupby('channel')['roi'].mean().reset_index()

fig = px.bar(roi_by_channel,
             x="channel", y="roi",
             title="Return on Investment by Channel")
fig.show()
```
